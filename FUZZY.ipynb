```python
import numpy as np
import skfuzzy as fz
from skfuzzy import control as ctrl
```

```python
# init agent

X1 = ctrl.Antecedent(np.arange(18,65,1),"Độ Tuổi")
X2 = ctrl.Antecedent(np.arange(0,3,0.1),"Trình độ Học Vấn")
X3= ctrl.Antecedent(np.arange(0,1,0.1),"Tình trạng Hôn nhân")
X4 = ctrl.Antecedent(np.arange(1,5,1),"Số con hợp pháp")
X5= ctrl.Consequent(np.arange(0,10,1),"Thông tin Nhân khẩu học")
```

```python
# age

X1["Trẻ"] = fz.trapmf(X1.universe,[18,20,27,33])
X1["Trung niên"] = fz.trapmf(X1.universe,[27,33,48,53])
X1["Già"] = fz.trimf(X1.universe,[48,53,65])
X1.view()
```

```python
# education level
X2["Thấp"] = fz.trimf(X2.universe,[0,0.8,1])
X2["Trung bình"] = fz.trapmf(X2.universe,[0.8,1,1.5,2.25])
X2["Cao"] = fz.trimf(X2.universe,[1.5,2.25,3])
X2.view()
```

```python
# marital status

X3["Độc thân"] = fz.trimf(X3.universe,[0,0,0.7])
X3["Có gia đình"] = fz.trimf(X3.universe,[0.7,1,1])
X3.view()
```

```python
# Number of child

X4["Ít"] = fz.trimf(X4.universe,[0,1,2])
X4["Vừa"] = fz.trimf(X4.universe,[1,2,3])
X4["Nhiều"] = fz.trimf(X4.universe,[3,5,7])
X4.view()
```

```python
from rembg import remove
from PIL import Image
import io
import matplotlib.pyplot as plt
```

```python
input_path = '/content/image_2023_07_11T08_24_49_052Z.png'
```

```python
# Open the image file
image_path = input_path
image = Image.open(image_path)

# Display the image
plt.imshow(image)
plt.axis('off')  # Hide the axis
plt.show()
```

```python
# replace input img = output img
output_path = input_path.replace("images", "results")

# Open the image file
with open(input_path, 'rb') as i:
    # Create a PIL Image object
    pil_image = Image.open(i)

    # Convert the image to RGBA format
    rgba_image = pil_image.convert('RGBA')

    # Create a BytesIO object to store the converted image
    image_stream = io.BytesIO()

    # Save the RGBA image to the stream in PNG format
    rgba_image.save(image_stream, format='PNG')

    # Retrieve the contents of the stream
    image_stream.seek(0)
    original_image = image_stream.read()

# Use rembg to remove the image background
output = remove(original_image)

# Write the output to the specified file
with open(output_path, 'wb') as o:
    o.write(output)
```

```python
# Open the image file
image_path = output_path
image = Image.open(image_path)

# Display the image
plt.imshow(image)
plt.axis('off')  # Hide the axis
plt.show()
```

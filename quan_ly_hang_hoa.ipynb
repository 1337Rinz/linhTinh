```python
class HangHoa:
    def __init__(self):
        self.Id = 0
        self.TenHangHoa = ""
        self.NuocSanXuat = ""
        self.NamSanXuat = 0
        self.TriGia = 0.0

    @property
    def ThueVAT(self):
        return self.TriGia * 0.1

    @property
    def KhuyenMai(self):
        return self.calculate_khuyen_mai()

    def calculate_khuyen_mai(self):
        if self.TenHangHoa == "Xe đạp Martin" and self.TriGia >= 1000000:
            return "Khuyến mãi: Xe đạp Martin"
        elif self.TenHangHoa == "Nón bảo hiểm" and 600000 <= self.TriGia < 1000000:
            return "Khuyến mãi: Nón bảo hiểm"
        elif self.TenHangHoa == "Bàn ủi" and 300000 <= self.TriGia < 600000:
            return "Khuyến mãi: Bàn ủi"
        else:
            return "Không khuyến mãi"


class QuanLyHangHoa:
    def __init__(self):
        self.danh_sach_hang_hoa = []

    def them_hang_hoa(self):
        print("Nhập thông tin hàng hoá:")
        while True:
            hang_hoa = HangHoa()

            hang_hoa.TenHangHoa = input("Tên hàng hoá: ")
            if hang_hoa.TenHangHoa.lower() == "e":
                break

            hang_hoa.NuocSanXuat = input("Nước sản xuất: ")
            hang_hoa.NamSanXuat = int(input("Năm sản xuất: "))
            hang_hoa.TriGia = float(input("Trị giá: "))

            self.danh_sach_hang_hoa.append(hang_hoa)

    def cap_nhat_hang_hoa(self):
        id_can_cap_nhat = int(input("Nhập ID của hàng hoá cần cập nhật: "))

        hang_hoa = next((hh for hh in self.danh_sach_hang_hoa if hh.Id == id_can_cap_nhat), None)
        if hang_hoa:
            print("Nhập thông tin cần cập nhật:")
            hang_hoa.TenHangHoa = input("Tên hàng hoá: ")
            hang_hoa.NuocSanXuat = input("Nước sản xuất: ")
            hang_hoa.NamSanXuat = int(input("Năm sản xuất: "))
            hang_hoa.TriGia = float(input("Trị giá: "))
            print("Cập nhật thông tin hàng hoá thành công!")
        else:
            print("Không có ID của hàng hoá trong danh sách.")

    def xoa_hang_hoa(self):
        id_can_xoa = int(input("Nhập ID của hàng hoá cần xoá: "))

        hang_hoa = next((hh for hh in self.danh_sach_hang_hoa if hh.Id == id_can_xoa), None)
        if hang_hoa:
            self.danh_sach_hang_hoa.remove(hang_hoa)
            print("Xoá hàng hoá thành công!")
        else:
            print("Không có ID của hàng hoá trong danh sách.")

    def tim_kiem_hang_hoa_theo_ten(self):
        ten_hang_hoa_can_tim = input("Nhập tên hàng hoá cần tìm kiếm: ")
        ket_qua = [hang_hoa for hang_hoa in self.danh_sach_hang_hoa if ten_hang_hoa_can_tim.lower() in hang_hoa.TenHangHoa.lower()]

        if ket_qua:
            print("Kết quả tìm kiếm:")
            for hang_hoa in ket_qua:
                print(f"ID: {hang_hoa.Id}, Tên hàng hoá: {hang_hoa.TenHangHoa}, Nước sản xuất: {hang_hoa.NuocSanXuat}, "
                      f"Năm sản xuất: {hang_hoa.NamSanXuat}, Trị giá: {hang_hoa.TriGia}, Thuế VAT: {hang_hoa.ThueVAT}, {hang_hoa.KhuyenMai}")
        else:
            print("Không tìm thấy hàng hoá phù hợp.")

    def hien_thi_danh_sach_hang_hoa(self):
        if self.danh_sach_hang_hoa:
            print("Danh sách hàng hoá:")
            for hang_hoa in self.danh_sach_hang_hoa:
                print(f"ID: {hang_hoa.Id}, Tên hàng hoá: {hang_hoa.TenHangHoa}, Nước sản xuất: {hang_hoa.NuocSanXuat}, "
                      f"Năm sản xuất: {hang_hoa.NamSanXuat}, Trị giá: {hang_hoa.TriGia}, Thuế VAT: {hang_hoa.ThueVAT}, {hang_hoa.KhuyenMai}")
        else:
            print("Danh sách hàng hoá trống.")


def main():
    quan_ly_hang_hoa = QuanLyHangHoa()

    while True:
        print("\n---- MENU QUẢN LÝ HÀNG HOÁ ----")
        print("1. Thêm Hàng hoá.")
        print("2. Cập nhật thông tin hàng hoá bởi ID.")
        print("3. Xóa Hàng hoá bởi ID.")
        print("4. Tìm kiếm hoá theo tên hàng hoá.")
        print("5. Hiển thị danh sách hàng hoá.")
        print("0. Thoát chương trình.")
        lua_chon = input("Vui lòng chọn chức năng: ")

        if lua_chon == "1":
            quan_ly_hang_hoa.them_hang_hoa()
        elif lua_chon == "2":
            quan_ly_hang_hoa.cap_nhat_hang_hoa()
        elif lua_chon == "3":
            quan_ly_hang_hoa.xoa_hang_hoa()
        elif lua_chon == "4":
            quan_ly_hang_hoa.tim_kiem_hang_hoa_theo_ten()
        elif lua_chon == "5":
            quan_ly_hang_hoa.hien_thi_danh_sach_hang_hoa()
        elif lua_chon == "0":
            print("Chương trình kết thúc.")
            break
        else:
            print("Chức năng không hợp lệ. Vui lòng chọn lại.")


if __name__ == "__main__":
    main()
```
